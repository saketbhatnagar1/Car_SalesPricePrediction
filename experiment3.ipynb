```python
import numpy as np
#Normalize value of engine, power, seats, price, age anf km driven
x = ["Engine", "Power", "Seats", "Price", "Age", "Kilometers_Driven", "Mileage"]
df1 = df
for i in x:
    df1[i] = df1[i]/np.abs(df1[i].max())
df.head()
```

```python
plt.boxplot(df1["Engine"])
```

```python
# create binary vector for fuel type
mydf = pd.DataFrame({'Gender':['M','F','M','F','F']})
print(mydf)
mydf1 = pd.get_dummies(mydf["Gender"])
mydf1

df2 = pd.get_dummies(df1["Fuel_Type"])
df2.shape
df2.head()
```

```python
df3 = pd.concat((df1,df2), axis=1)
df3.head()
```

```python
# create binary vector for transmission
df4 = pd.get_dummies(df["Transmission"])
df4.head()
```

```python
df5 = pd.concat((df3,df4), axis=1)
df5.head()
```

```python
# delete fuel type and transmission
df5.drop(columns=["Fuel_Type","Transmission"],axis=1,inplace=True)
df5.head()
```

```python
#convert owner type into numeric value
df5["Owner_Type"].value_counts()

dic_owner = {"First":2, "Second and above":1}
df5["Owner_Type"] = df5["Owner_Type"].map(dic_owner)
df5.head()
```

```python
# find outliers for all numeric attributes
plt.boxplot(df5["Engine"])
plt.xlabel("Engine")

```

```python
# detect outliers from age and replace with mean value
q = list(df5["Age"].quantile([0,0.25,0.5,0.75,1]))
q
```

```python
df5["Age"].median()
df5["Age"].mean()
df5["Age"].max()
```

```python
IQR = q[3] - q[1]
IQR
```

```python
LIF = q[1] - 1.5*IQR
LOF = q[1] - 3*IQR
UIF = q[3] + 1.5*IQR
UOF = q[3] + (3*IQR)

upper_outlier = df5[df5["Age"] >= UIF].index
print(upper_outlier)
```
