```python
def fill_mileage(x):
    if x == 'Diseal':
        return 18.778682
    elif x == 'Petrol':
        return 17.679168
    else:
        return 24.503939
fill_mileage("Petrol")
```

```python
df_mil0["Mileage"] = df_mil0["Fuel_Type"].apply(fill_mileage)
```

```python
df_mil0.head()
```

```python
df1= pd.concat((df_mil0,df_mil1))
(df1 == 0).sum()
```

```python
#eliminate zeroes from Engine
df_eng0 = df1[df1["Engine"]==0]
df_eng1 = df1[df1["Engine"]!=0]
```
